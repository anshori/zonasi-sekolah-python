# Zonasi Sekolah

```python
import pandas as pd
import os

# Sekolah SMPN 16 Yogyakarta
school_coordinates = (-7.8113051,110.3592539)

data_pkg_path = 'data'
filename = 'siswa_baru.csv'
path = os.path.join(data_pkg_path, filename)

# list data siswa baru (dsb)
dsb = pd.read_csv(path)
dsb
```

```python
from geopy import distance

# calculate distance from lintang & bujur siswa to school coordinates
def calculate_distance(row):
    siswa_coordinates = (row['lintang'], row['bujur'])
    return distance.geodesic(siswa_coordinates, school_coordinates).m

result = dsb.apply(calculate_distance, axis=1)
print(result)
```

```python
# merge calculate distance to data siswa
dsb['distance_meter'] = result

# sort data siswa by distance
dsb = dsb.sort_values(by='distance_meter')
print(dsb)

# save data siswa to csv
output_filename = 'siswa_baru_distance.csv'
output_dir = 'output'
output_path = os.path.join(output_dir, output_filename)
dsb.to_csv(output_path, index=False)
```
